# Comparación Agroclimática entre Soja, Maíz y Trigo

## Análisis comparativo de sensibilidad climática y desempeño predictivo

En este notebook se realiza una comparación integral entre los modelos agroclimáticos desarrollados para:

- soja,
- maíz,
- y trigo.

El objetivo es analizar diferencias en:

- sensibilidad climática,
- variables atmosféricas dominantes,
- impacto de eventos extremos,
- estabilidad temporal,
- y capacidad predictiva.

## Motivación

Cada cultivo presenta respuestas fisiológicas y climáticas diferentes frente a:

- temperatura,
- precipitación,
- humedad atmosférica,
- estrés evaporativo,
- y eventos extremos.

Por este motivo, resulta necesario construir un análisis comparativo que permita comprender cómo responde cada sistema agrícola frente a distintos escenarios climáticos.

## Objetivos específicos

- Comparar métricas predictivas entre cultivos.
- Analizar diferencias en importancia de variables.
- Evaluar estabilidad temporal mediante backtesting.
- Comparar sensibilidad frente a eventos extremos.
- Identificar patrones agroclimáticos dominantes.

## Hipótesis general

Los resultados previos sugieren que:

- el maíz presenta fuerte dependencia hídrica,
- la soja posee una respuesta intermedia,
- y el trigo muestra elevada sensibilidad atmosférica y evaporativa.

Este notebook permitirá validar y sintetizar dichas diferencias dentro de un marco agroclimático comparativo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)

import warnings
warnings.filterwarnings("ignore")
```

## Resumen comparativo de desempeño predictivo

Se construye una tabla comparativa con las principales métricas obtenidas para cada cultivo en sus modelos finales.

```python
comparacion_modelos = pd.DataFrame({

    "Cultivo": [
        "Soja",
        "Maíz",
        "Trigo"
    ],

    "Modelo_final": [
        "RF agroclimático avanzado",
        "RF ENSO + VPD + humedad",
        "RF + eventos extremos"
    ],

    "R2_RF": [
        None,   # completar con soja
        0.901,
        0.919
    ],

    "MAE_RF": [
        None,   # completar con soja
        490.04,
        205.94
    ],

    "RMSE_RF": [
        None,   # completar con soja
        699.50,
        285.40
    ]

})
```

```python
comparacion_modelos
```

## Interpretación comparativa inicial

Los resultados muestran que los modelos agroclimáticos desarrollados lograron elevados niveles de capacidad predictiva para maíz y trigo.

Particularmente:

- el trigo presentó el mayor desempeño global,
- mientras que el maíz mostró una elevada sensibilidad frente a variables atmosféricas avanzadas como humedad relativa, VPD y ENSO.

Los resultados sugieren que ambos cultivos poseen respuestas climáticas complejas y altamente dependientes de la dinámica atmosférica.

La soja será incorporada posteriormente para completar el análisis multicultivo final.

## Comparación gráfica de desempeño predictivo

Se comparan visualmente los valores de R² obtenidos para cada cultivo con el objetivo de evaluar diferencias en capacidad predictiva entre modelos agroclimáticos.

```python
plt.figure(figsize=(8,5))

plt.bar(

    comparacion_modelos["Cultivo"],
    comparacion_modelos["R2_RF"]

)

plt.title(
    "Comparación de R² - Modelos Agroclimáticos"
)

plt.ylabel("R²")

plt.show()
```

## Limitaciones del estudio

Aunque los modelos desarrollados mostraron una elevada capacidad predictiva y permitieron identificar patrones agroclimáticos relevantes, existen algunas limitaciones importantes.

## Principales limitaciones

### Agregación temporal anual

Las variables climáticas fueron agregadas por campaña agrícola, lo que puede suavizar señales críticas asociadas a:

- eventos extremos intraestacionales,
- olas de calor,
- heladas,
- y períodos secos prolongados.

### Resolución espacial

El análisis climático se basó principalmente en datos regionales agregados, lo que puede no representar completamente variaciones microclimáticas locales.

### Variables fenológicas

El modelo no incorpora etapas fenológicas específicas de cada cultivo, como:

- floración,
- llenado de grano,
- o períodos críticos de estrés.

### Campañas fuera de régimen

Las campañas climáticamente extremas mostraron importantes limitaciones de generalización temporal, indicando la existencia de dinámicas atmosféricas complejas difíciles de representar mediante modelos agregados.

## Consideraciones metodológicas

Los resultados obtenidos deben interpretarse como aproximaciones agroclimáticas regionales y no como sistemas determinísticos exactos de predicción agrícola.
